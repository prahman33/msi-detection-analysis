# Microsatellite Instability (MSI) Analysis 🧬

This notebook explores patterns in simulated cancer genomic data to understand microsatellite instability (MSI).

Author: Mahbuba Rahman Prianka

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
# Simulated dataset for MSI analysis

np.random.seed(42)

data = pd.DataFrame({
    'Sample_ID': range(1, 101),
    'MSI_Score': np.random.normal(loc=10, scale=3, size=100),
    'Mutation_Count': np.random.poisson(lam=20, size=100),
    'Tumor_Type': np.random.choice(['MSI-High', 'MSI-Low'], size=100)
})

data.head()
```

```python
print("Dataset Info:")
print(data.info())

print("\nSummary Statistics:")
print(data.describe())
```

```python
print("Dataset Info:")
print(data.info())

print("\nSummary Statistics:")
print(data.describe())
```

```python
plt.figure()
plt.hist(data['MSI_Score'], bins=15)
plt.title("Distribution of MSI Scores")
plt.xlabel("MSI Score")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure()
sns.boxplot(x='Tumor_Type', y='Mutation_Count', data=data)
plt.title("Mutation Count by Tumor Type")
plt.show()
```

```python
correlation = data[['MSI_Score', 'Mutation_Count']].corr()
print(correlation)

plt.figure()
sns.heatmap(correlation, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure()
sns.scatterplot(x='MSI_Score', y='Mutation_Count', hue='Tumor_Type', data=data)
plt.title("MSI Score vs Mutation Count")
plt.show()
```

```python
## 🔍 Key Insights

- MSI-High tumors tend to show higher mutation counts
- Positive correlation observed between MSI score and mutation count
- Data visualization helps identify patterns in genomic instability

This demonstrates how computational analysis can support cancer genomics research.
```
